```python
# Importando bibliotecas
import pandas as pd

# Variavel para apontar o arquivo original
file_path = '../data/raw/BASE DE DADOS PEDE 2024 - DATATHON.xlsx'

# Ler o arquivo e apontar as abas que existem
xls = pd.ExcelFile(file_path)

nomes_das_abas = xls.sheet_names

print(nomes_das_abas)
```

```python
# Carregando as abas em dataframes separados
df_2022 = pd.read_excel(file_path, sheet_name='PEDE2022')
df_2023 = pd.read_excel(file_path, sheet_name='PEDE2023')
df_2024 = pd.read_excel(file_path, sheet_name='PEDE2024')

# Exibindo tamano de linhas e colunas de cada dataframe
print(f"PEDE2022: {df_2022.shape}")
print(f"PEDE2023: {df_2023.shape}")
print(f"PEDE2024: {df_2024.shape}")

# Exibindo nome das colunas de cada dataframe
print(f"\nColunas PEDE2022: {df_2022.columns.tolist()}")
print(f"Colunas PEDE2023: {df_2023.columns.tolist()}")
print(f"Colunas PEDE2024: {df_2024.columns.tolist()}")

# Exibindo informações gerais dos dataframes
print("\nInformações PEDE2022:")
print(df_2022.info())
print("\nInformações PEDE2023:")
print(df_2023.info())
print("\nInformações PEDE2024:")
print(df_2024.info())
```

```python
# Renomeando colunas para padronização
df_2022.rename(
    columns={
        'Nome': 'Nome Anonimizado',
        'Idade 22': 'Idade',
        'Matem': 'Mat',
        'Portug': 'Por',
        'Inglês': 'Ing',
        'Defas': 'Defasagem',
        'INDE 22': 'INDE',
        'Pedra 22': 'Pedra',
        'Fase ideal': 'Fase Ideal'
    },
    inplace=True
)

df_2023.rename(
    columns={
        'INDE 2023': 'INDE',
        'Pedra 2023': 'Pedra'
    },
    inplace=True
)

df_2024.rename(
    columns={
        'INDE 2024': 'INDE',
        'Pedra 2024': 'Pedra'
    },
    inplace=True
)

# Removendo colunas desnecessárias ou sem dados
df_2022.drop(
    columns=['Cg', 'Cf', 'Ct', 'Rec Av1', 'Rec Av2', 'Rec Av3', 'Rec Av4', 'Destaque IEG', 'Destaque IDA', 'Destaque IPV', 'Indicado', 'Atingiu PV', 'Rec Psicologia', 'Avaliador1', 'Avaliador2', 'Avaliador3', 'Avaliador4', 'Pedra 20', 'Pedra 21'],
    inplace=True
)

df_2023.drop(
    columns=['Cg', 'Cf', 'Ct', 'Rec Av1', 'Rec Av2', 'Rec Av3', 'Rec Av4', 'Destaque IEG', 'Destaque IDA', 'Destaque IPV', 'Destaque IPV.1', 'Indicado', 'Atingiu PV', 'Rec Psicologia', 'Avaliador1', 'Avaliador2', 'Avaliador3', 'Avaliador4', 'Pedra 20', 'Pedra 21', 'Pedra 22', 'Pedra 23', 'INDE 22', 'INDE 23'],
    inplace=True
)

df_2024.drop(
    columns=['Cg', 'Cf', 'Ct', 'Rec Av1', 'Rec Av2', 'Destaque IEG', 'Destaque IDA', 'Destaque IPV', 'Indicado', 'Atingiu PV', 'Rec Psicologia', 'Avaliador1', 'Avaliador2', 'Avaliador3', 'Avaliador4', 'Avaliador5', 'Avaliador6', 'Pedra 20', 'Pedra 21', 'Pedra 22', 'Pedra 23', 'Ativo/ Inativo.1', 'INDE 22', 'INDE 23'],
    inplace=True
)

# Adicionando coluna de ano para cada dataframe
df_2022['Ano'] = 2022
df_2023['Ano'] = 2023
df_2024['Ano'] = 2024

# Concatenando os dataframes em um único dataframe
df_historico = pd.concat([df_2022, df_2023, df_2024], ignore_index=True)

df_historico.shape
```

```python
df_historico.info()
```

```python
# Limpando texto da coluna INDE e convertendo para numérico
df_historico['INDE'] = pd.to_numeric(df_historico['INDE'], errors='coerce')

# Imprimindo os valores unicos da coluna pedra e contando a frequencia de cada valor
print(df_historico['Pedra'].unique())
```

```python
# substituindo valor da pedra "Incluir" por vazio
df_historico['Pedra'] = df_historico['Pedra'].replace('INCLUIR', pd.NA)
```

```python
df_historico.info()
```

```python
# Tratando a coluna de idade
# pegar somente as linhas que contem '1900'
mask_erro = df_historico['Idade'].astype(str).str.contains('1900')

# transformando o erro em data e extraindo somente o dia
df_historico.loc[mask_erro, 'Idade'] = pd.to_datetime(df_historico.loc[mask_erro, 'Idade']).dt.day

# convertendo a coluna de idade para numérico
df_historico['Idade'] = pd.to_numeric(df_historico['Idade'], errors='coerce')
```

```python
# Importanto biblioteca para visualização
import matplotlib.pyplot as plt
import seaborn as sns

# Boxplot para verificar distribuição das notas ao longo do tempo
plt.figure(figsize=(14, 5))
sns.boxplot(data=df_historico, x='Ano', y='INDE', palette='colorblind', hue='Ano')
plt.title('Distribuição do INDE ao longo dos anos')
plt.xlabel('Ano')
plt.ylabel('Nota INDE')
plt.tight_layout()
plt.show()
```

```python
# Criando Heatmap (mapa de calor das correlações)
# Selecionando apenas colunas de correlação
colunas_correlacao = ['INDE', 'IDA', 'IEG', 'IAA', 'IPS', 'IPP', 'IPV']

# Calculando a matriz de correlação
matriz_correlacao = df_historico[colunas_correlacao].corr()

# Plotando o heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(matriz_correlacao, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Matriz de Correlação entre as Variáveis')
plt.tight_layout()
plt.show()
```
